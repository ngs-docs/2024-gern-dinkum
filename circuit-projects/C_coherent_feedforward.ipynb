```python
import dinkum
from dinkum.vfg import Gene, Receptor
from dinkum.vfn import Tissue
```

```python
dinkum.reset()

ti1= Gene(name="Ti1")
ti2= Gene(name="Ti2")
ti3= Gene(name="Ti3")
a = Gene(name='A')
b = Gene(name='B')
c = Gene(name='C')
d = Gene(name='output')
m = Tissue(name='M')

ti1.is_present(where=m, start=1, duration=1)
ti2.is_present(where=m, start=1, duration=1)
ti3.is_present(where=m, start=1, duration=1)
a.activated_by(source=ti1, delay=2)
b.activated_by_and(sources=[ti2, a], delay=2)
c.activated_by_and(sources=[ti3, a, b], delay=2)
d.activated_by(source=c, delay=2)

dinkum.run_and_display(stop=10, gene_names=['Ti1',  'A', 'Ti2', 'B', 'Ti3', 'C', 'output'])
```
