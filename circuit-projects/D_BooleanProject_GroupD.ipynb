# Incoherent feed forward circuit

```python
import dinkum
from dinkum.vfg import Gene, Receptor
from dinkum.vfn import Tissue
```

```python
dinkum.reset()

#define our genes and tissues
sca = Gene(name="SCA")
bda = Gene(name="ubiq")
a_1 = Gene(name="A1")
a_2 = Gene(name="A2")
b_1 = Gene(name="B1")
c = Gene(name="C")

m = Tissue(name='M')
n = Tissue(name='N')

#define a regulatory relationships
bda.is_present(where=m, start=1, duration=10)
bda.is_present(where=n, start=1, duration=10)
sca.is_present(where=n, start=1, duration=10)
a_1.activated_by(source=bda, delay=1)
a_2.activated_by(source=bda, delay=1)
b_1.activated_or(sources=[a_1, sca], delay=3)
c.and_not(activator=a_1, repressor=b_1, delay=1)

#run it!
dinkum.run_and_display(stop= 10, gene_names=['ubiq', 'SCA', 'A1', 'A2', 'B1', 'C'])
```
